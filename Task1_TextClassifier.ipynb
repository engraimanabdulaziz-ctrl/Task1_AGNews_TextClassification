```python
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
dataset = load_dataset("ag_news")

train_texts = dataset["train"]["text"]
train_labels = dataset["train"]["label"]

test_texts = dataset["test"]["text"]
test_labels = dataset["test"]["label"]

label_names = dataset["train"].features["label"].names
print("Labels:", label_names)
```

```python
vectorizer = TfidfVectorizer(max_features=5000)  # top 5000 words
X_train = vectorizer.fit_transform(train_texts)
X_test = vectorizer.transform(test_texts)
```

```python
clf = LogisticRegression(max_iter=1000)  # Initialize the Logistic Regression model
clf.fit(X_train, train_labels)  # Train the model
```

```python
def predict_headline(text):
    X_new = vectorizer.transform([text])
    pred = clf.predict(X_new)[0]
    return label_names[pred]

print(predict_headline("Microsoft announces new AI breakthrough"))
print(predict_headline("Pakistan wins T20 cricket match"))
```
